```python
# Implementing simple neuron
import numpy as np
class Neuron:
    def __init__(self,inputs,weight,bias):
        self.input=inputs
        self.weight=weight
        self.bias=bias
        self.output=None
    def forward_pass(self):
        self.output=np.dot(self.input,self.weight)+self.bias
        return self.output
    def backward_pass(self,yh,y,learning_rate):
        dE_dyh=yh-y
        dw1=self.input[0]*dE_dyh
        dw2=self.input[1]*dE_dyh
        db=dE_dyh
        
#         Updating parameters
        self.weight[0]-=learning_rate*dw1
        self.weight[1]-=learning_rate*dw2
        self.bias-=learning_rate*db
#         self.weight[1]=self.weight[1]-learning_rate*dw2
#         self.bias=self.bias-learning_rate*db
# Trainging neuron with different inputs
# data=np.array([[2,3],[3,3],[2,2.3],[4,1],[2,2]])
# tvalue=np.array([9,12,8.5,13,8])
N=Neuron([2,3],[0.5,1.2],[1])
# print(N.forward_pass())
err=0
for epoches in range(63):
#     N.input=data[i]
    yh=N.forward_pass()
#     err+=(yh-tvalue[i])**2
    N.backward_pass(yh,9,0.02)
    print(yh)
# print(err/5)
```
